```python
#Since EmployeeCount(all numbers are 1), EmployeeNumber(sequential count) and StandardHours(all values are 80) columns have very little info, we can remove them
num_cols = list(df.describe().columns) #usinf describe we get only the numerical columns and .columns gives us the names of the numerical columns
cat_cols = list(df.columns.difference(num_cols))
remove_cols = ['EmployeeCount', 'EmployeeNumber', 'StandardHours']
cols = [c for c in num_cols if c not in remove_cols]
```

```python
cat_cols
```

```python
print(len(num_cols))
print(len(cat_cols))
print(len(cols))
```

```python
plt.figure(figsize=(24,16))
sns.heatmap(df[cols].corr(), annot=True,fmt = ".2f")
```
